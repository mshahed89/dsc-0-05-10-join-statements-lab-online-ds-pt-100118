## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
#Your code here
import pandas as pd
import sqlite3
connection = sqlite3.connect('data.sqlite', detect_types=sqlite3.PARSE_COLNAMES)
cursor = connection.cursor()
```

## Display the names of all the employees in Boston.

```python
#Your code here
cursor.execute(""" select firstname,lastname from employees e
join offices o on o.officecode=e.officecode 
where city ="Boston"
;
""").fetchall()
```

## Do any offices have no employees?

```python
#Your code here
cursor.execute("""
select city,count(*) from offices o
left join employees e on o.officecode=e.officecode 
group by city
;
""").fetchall()
```

## Level Up: Display the names of each product each employee has sold.

```python
# Your code here
cursor.execute("""
select e.firstname,e.lastname,p.productname from employees e
join customers c on c.salesrepemployeenumber = e.employeenumber
join orders o on o.customernumber = c.customernumber
join orderdetails od on od.ordernumber = o.ordernumber
join products p on p.productcode = od.productcode
;
""").fetchall()
```

## Level Up: Display the Number of Products each Employee Has sold

```python
#Your code here
cursor.execute("""
select e.firstname,e.lastname,count(p.productcode) from employees e
join customers c on c.salesrepemployeenumber = e.employeenumber
join orders o on o.customernumber = c.customernumber
join orderdetails od on od.ordernumber = o.ordernumber
join products p on p.productcode = od.productcode
group by e.firstname,e.lastname
;
""").fetchall()
```
